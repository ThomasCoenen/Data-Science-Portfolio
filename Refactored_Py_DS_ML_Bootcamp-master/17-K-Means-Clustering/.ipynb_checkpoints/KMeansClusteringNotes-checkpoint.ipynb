How KMeans Algorithm Works
 -Choose a number of Cluserts 'K'
 -Randomly assign each point to a cluster
 -Until clusters stop changing repeat the following:
   -for each cluster compute the cluster centroid by taking the mean vector of points in the cluster
   -assing each data point to the cluster for which the centroid is the closest
 
#Choosing a K-VALUE
We need to decide how many CLUSTER we expect in the data. The problem of selection K isnt that simple - no easy answer.
One way to do it is use the ELBOW METHOD
For ELBOW METHOD - compute the Sum of Squared Error(SSE) for some values of K
The SSE is defined as the sum of the squared distance between each member of the cluster and its centroid

#If you plot K against SSE - ERROR DECREASES as K gets LARGER, this is bc when the number of clusters increases they should be smaller so distortion is also smaller. 
#The idea of the ElbowMethod is to choose the K at thich the SSE DECREASES ABRUPTLY
#This produces an ELBOW EFFECT in the graph 

#So we choose a K where we wont get that much more information by increasing number of Clusters - meaning we are not going to signifiantly decrease the within groups sum of squared error by increasing clusters

#THATS WHY its called ELBOW METHOD. At the ELBOW this is where we choose clusters

#THERE IS NO CORRECT K VALUE SO USE FIELD AND DOMAIN KNOWLEDGE WHERE WE WORK TO CHOOSE RIGHT K VALUE

```python
# K Means with Python
#We will learn how to use algorith and also learn how
#to create ARTIFICIAL DATA using SK LEARNS 
 #capabilities

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#USE SK LEARN to generate ARTIFICIAL DATA
from sklearn.datasets import make_blobs
#This dataset generates blob-like groups of data

#Create Data
#200 samples, 2 Features(so 2 Cols),Centers=4(so 4 
  #BLOBS of data)
df = make_blobs(n_samples=200,n_features=2,centers=4,
                cluster_std=1.8,random_state=101)
```

```python
#Explose data
#our data is a TUPLE!
df
```

```python
df[0]
#This gives up a NUMPY ARRAY. This Numpy Array is your
#number of samples and 2 Cols of FEATURES
```

```python
df[0].shape
```

```python
#Lets Plot Blobs 
##grab all rows in First Col and plot against all 
 #rows in Second Col
plt.scatter(df[0][:,0],df[0][:,1])
#We see 2 blobs so we need to grab the SECOND ITEM
 #in the TUPLE
```

```python
#grab the SECOND ITEM in the TUPLE
df[1]
#this is the CLUSTER that it belongs to so 0,1,2,3 and
#thats from the CENTER. So theres 4 centers to these 
#blobs
```

```python
#SPECIFY 'C'
plt.scatter(df[0][:,0],df[0][:,1],c=df[1],cmap='rainbow')
#So this is how yo usue SKLearn to create Artificial 
#Data for CLustering
```

```python
#Create KMEANS Clusting Algorithm
#Remember KMEANS is Unsupervised. So it will randomly 
# assigm each observation to a cluster and finds 
# CENTROID of cluster. The algorithm will iterate thru 
# the 2 steps, it reassigns data pts to the cluster 
# whos centroid is closest, calculates new centroid for 
# each cluseter and repeasts over and over until cluster
# vairationcant be reduces any further

from sklearn.cluster import KMeans

#Create KMEANS OBJECT
#In this case we KNOW the clusters beforehand since 
 #we created data and set it to 4
kmeans = KMeans(n_clusters=4)
```

```python
#FIT DATA to the FEATURES and the FEAUTURES are the 
 #df[0]
kmeans.fit(df[0])
```

```python
#Get CENTERS of CLUSTERS
kmeans.cluster_centers_
#Get array of CENTERS of CLusters (4)
#Notice the Number of 
#dimensions. This is the exact same number of 
#dimensions as you had FEAUTUES in your dataset
```

```python
#GET PREDICTED LABELS of CLUSTERS
#So this is what the ALGOIRTHM thinks the LABELS 
 #SHOULD BE
#This will give us the PREDICTED LABELS it BELIVES 
 #TO BE TRUE for the CLUSTERS
kmeans.labels_
#Something importatnt to keep in mind, if you were
# working with real data and you didn't have labels you
# would be done at this stage. You wouldnt be able to 
# actually compare anything to the REAL VALUES. But 
# we acutally know the real labels in this case.

#Again K Means is Unsupervised so you wouldnt use it to 
#predict labels you would use it to try and find labels
#in your data.

#YOu would do that by exploring number of Clusters

```

```python
#How does KMEANS compare to our original? Again
#  in real world we would be done but since we created
#     the data we can actually compare it

#Use MPL
#CREATET SUBPLOT 1ROW by 2COLUMNS
#Sharey=True: means they SHARE an axis
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True,figsize=(10,6)) 
#Using Tuple Unpacking to GRAB the 2 set of AXIS

# First set of axes plots the original data and colors 
# data based of what algorithm thought labels should 
# look like. Second plot Ax2 plots ORIGINAL DATA and 
# COLORS them by their FACTUALLY CORRECT BLOG 
# ASSIGNMENTS


#grab all rows in First Col and plot against all 
 #rows in Second Col (just like before)
#C = Color (which are our KMEANS LABEL CLUSTER 
  #PREDICITONS )
ax1.set_title('K Means')
ax1.scatter(df[0][:,0],df[0][:,1],c=kmeans.labels_,
            cmap='rainbow')

#C=COLOR CLUSTER where Original Data belong to
ax2.set_title('Original Data')
ax2.scatter(df[0][:,0],df[0][:,1],c=df[1],
            cmap='rainbow')
#Coloring is differnet but dont wory about that too 
# much. We see in the NOISY middle section is wehre our
# algorithm couldnt detect which point belonged to which 
# group. The coloring is meaningless and just colors
#the differnet groups
```

```python
#What if we set KMEANS=2?

#Make Object
kmeans = KMeans(n_clusters=2)

#FIT DATA to the FEATURES and the FEAUTURES are the 
 #df[0]
kmeans.fit(df[0])

#Get CENTERS of CLUSTERS
kmeans.cluster_centers_ 
 #get 2 Cluster Centers. Notice the Number of 
#dimensions. This is the exact same number of 
#dimensions as you had FEAUTUES in your dataset
```

```python
#Get Predicted Labels
kmeans.labels_
```

```python
#Plot agin with Kmeans=2
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True,figsize=(10,6)) 
ax1.set_title('K Means')
ax1.scatter(df[0][:,0],df[0][:,1],c=kmeans.labels_,
            cmap='rainbow')

#C=COLOR CLUSTER where Original Data belong to
ax2.set_title('Original Data')
ax2.scatter(df[0][:,0],df[0][:,1],c=df[1],
            cmap='rainbow')
```
