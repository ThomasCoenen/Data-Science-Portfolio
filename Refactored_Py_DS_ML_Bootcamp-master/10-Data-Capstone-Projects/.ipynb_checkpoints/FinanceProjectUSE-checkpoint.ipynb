```python
import pandas as pd

import datetime

from pandas import Series,DataFrame

import pandas_datareader
import pandas_datareader.data as web

pandas_datareader.__version__

start=datetime.datetime(2006, 1, 1)

end=datetime.datetime(2016, 1, 1)

bac=web.DataReader( 'BAC', "yahoo", start,end) # bank of america

print(bac.head())
```

```python
# CitiGroup
cg=web.DataReader( 'C', "yahoo", start,end)

# Goldman Sachs
gs=web.DataReader( 'GS', "yahoo", start,end)

# JPMorgan Chase
jpm=web.DataReader( 'JPM', "yahoo", start,end)

# Morgan Stanley
ms=web.DataReader( 'MS', "yahoo", start,end)

# Wells Fargo
wfc=web.DataReader( 'WFC', "yahoo", start,end)
```

** Create a list of the ticker symbols (as strings) in alphabetical 
order. Call this list: tickers**

```python
tickers = ['bac', 'cg', 'gs', 'jpm', 'ms', 'wfc']
```

** Use pd.concat to concatenate the bank dataframes together to a single 
data frame called bank_stocks. Set the keys argument equal to the 
tickers list. Also pay attention to what axis you concatenate on.**

```python
# Concat along columns (axis=1) - we can concat them since they all
 #share the same start and end datetime index
bank_stocks = pd.concat([bac,cg,gs,jpm,ms,wfc],axis=1,keys=tickers)
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
# So this will give us a percent return from previous day
# rt = return for some time period
# (price @ time t/previous date) - 1 
# (price at a date/ previous date) - 1  = percent return
```

```python
#This will contain returns for each banks stock
returns = pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
bank_stocks.head(2)
```

```python
# create a returns column ( we added new DF for this)
  # and create a new column - Column name will be tick
  # concatenated with return = bank stocks at that tick
  # closing price.pct change
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
```

** Take a look at the standard deviation of the returns, which stock would you classify as the riskiest over the entire time period? Which would you classify as the riskiest for the year 2015?**
