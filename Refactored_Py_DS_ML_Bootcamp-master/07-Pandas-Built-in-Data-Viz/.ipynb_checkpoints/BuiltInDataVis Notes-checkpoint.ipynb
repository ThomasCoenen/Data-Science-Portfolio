```python
# Pandas has several plots built in
#Most of the plots can be made off a command off the DF in
 # 2 Methods: 1 - call the DF itself or a column off DF
            # 2- instead of specifying kind=hist, we can
                  #call hist directly off of plot
```

```python
# Method 1 - call the DF itself or a column off DF
df1['A'].plot(kind='hist',bins=30)
```

```python
# Method 2 - call hist directly off of plot
df1['A'].plot.hist(bins=30)
```

```python
df1.head()
```

```python
#SCATTER PLOT - Set COLOR to show another Column 'C'
 #for instance we want color (c) to be equal off column C
# So when looking at head we have values for A,B, and C
 # so we are now showing 3 levels of INFO A vs B and a
 # third element C - shows the Color Value for C
 #CMAP - to fix colors
df1.plot.scatter(x='A',y='B',c='C')
#So we get a black and white
```

```python
#SCATTER PLOT - CMAP
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
#SCATTER PLOT - Show 'C' by SIZE
# Some people don't like to show things by color but by 
  #SIZE - pass in S as ARGUMENT - must pass in DF column
df1.plot.scatter(x='A',y='B',s=df1['C'])
  # DO THIS POINTS are too SMALL - multiply by a factor
```

```python
#SCATTER PLOT - Show 'C' by SIZE - MULTIPLY by FACTOR
#So now points are A vs B and SIZE indicates how large
 # C value was relative to each other
df1.plot.scatter(x='A',y='B',s=df1['C']*30)
```

```python
# BIVARIATE DATA
  #for Bivariate data do a HEXAGONAL BIN PLOT and we saw 
    # this before with Seaborn
  # Make new DF below for this to work correctly 
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
df.shape
```

```python
df.head()
```

```python
#HEXBIN
 # use for 2 variables
df.plot.hexbin(x='a',y='b')
# If we just do this we get a HEXPLOT and essentially a
#   SCATTERPLOT except get the HEXAGONAL BINS
# BINS are small - change the gridsize
```

```python
#HEXBIN - change GRIDSIZE and CMAP
 #GRIDSIZE will enlarge size of the Hexagons
 #Hexagons get darker as there are more points inside of 
    #them
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='coolwarm')
```
