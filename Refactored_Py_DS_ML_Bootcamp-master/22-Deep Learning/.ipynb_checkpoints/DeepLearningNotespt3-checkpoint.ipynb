#How to perform a CLASSIFICATION Task with TensorFlow
How to Identify and deal with OVERFITTING thru EARLY STOPPING CALLBACKS and DROPOUT LAYERS

#EARLY STOPPING 
#Keras can automatically STOP TRAINING based on a LOSS condition on the VAL DATA passed during the MODEL.FIT call

#DROPOUT LAYERS
 -DROPOUT can be added to layers to 'TURN OFF' Nurons during TRAINING to prevent OVERFITTING
 -Each DROPOUT LAYER will DROP a user defined PERCENTAGE of NURON units in the previous layer every batch. So this means certain Nurons dont have their WEIGHTS or BIASES affected during a BATCH instead they are just TURNED OFF

```python
#TRAIN/TEST SPLIT

from sklearn.model_selection import train_test_split

#Grab VALUES so its a NUMPY ARRAY
#FEATURES 
X = df.drop('benign_0__mal_1',axis=1).values

#LABEL
y = df['benign_0__mal_1'].values

#TRAIN/TEST SPLIT
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
#SCALE DATA
from sklearn.preprocessing import MinMaxScaler

#Make INSTANCE of SCALER
scaler = MinMaxScaler()

#FIT/TRANSFORM TRAINING DATA
X_train = scaler.fit_transform(X_train)

#FIT/TRANSFORM TEST DATA
#Recall we dont want to FIT to TEST DATA we only FIT to 
#TRAINING data to prevent data leakage
X_test = scaler.transform(X_test)
```

```python
#NEXT we will CREATE MODEL and learn how to prevent 
# OVERFITTING and how make sure you dont OVERRUN on the 
# TRAINING DATA and end up having a poor fit on TEST DATA
```

```python
#TF Classification - Dealingwith Overfitting and Evaluation
#CREATE MODEL and PREVENT OVERFITTING
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
```

```python
X_train.shape
```

```python
#CREATE MODEL
# Right now in our TRAINING SET we ahve 426 rows and 30 
# FEAUTURES

model = Sequential()

#1st Layer: 30 Nurons
model.add(Dense(30,activation='relu'))
#2nd Layer: 15 Nurons
model.add(Dense(15,activation='relu'))

#OUTPUT LAYER: 1 NURON
 #ACTIVATIOJN(and this is critical)=SIGMOID bc this is a 
    #BINARY CLASSIFICATION use SIGMOID
#So we have BINARY CLASSIFCATION, 1 NURON outputing 
# something between a 0 and 1 which will decide what CLASS 
# it belongs to (MAL or BEN) 
model.add(Dense(1,activation='sigmoid'))

#COMPILE Model
#Make sure LOSS=BINARY CROSSENTROPY
#OPTIMIZER
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
#TRAIN MODEL
#EPOCHS trains 600 times thru entire Training set
#Choose REALLY large EPOCHS(probly too many). BC we want to
#see what it looks like when we OVERFIT to TRAINING Data
model.fit(x=X_train,y=y_train,epochs=600,
                validation_data=(X_test,y_test))
```

```python
#PLOT out LOSS
#Remeber we pass in VAL DATA so we can PLOT both TRIANING
#LOSS and VAL LOSS
model.history.history
```

```python
losses = pd.DataFrame(model.history.history)
losses
```

```python
#Plot LOSSES
losses.plot()
#PERFECT EXAMPLE OF OVERFITTING
#You'll notice in Beginning both VAL and TRAIN LOSS and both
# decressing which is good and means we haven't yet OVERFIT 
# to our TRIANING data and as we go along we are decreasing
# loss in both VAL Set and TRAIN SET however at a certain 
# point in time notice that our TRIANING loss in blue is still
# going down, hoever our VAL LOSS is beginning to increase 
# and that tells us we are overfittign to TRAINING data set
```

```python
#So we are TRAINING for TOO MANY EPOCHS and VAL gets WORSE
 #and WORSE
#So we will see if we an use EARLY STOPPING and use TF KERAS
# CALLBACKS to actually based on your VAL LOSS, STOP the 
# TRIANING before it gets out of hand
```

```python
#RECREATE MODEL (same as above havent changed anything yet)
model = Sequential()

#1st Layer: 30 Nurons
model.add(Dense(30,activation='relu'))
#2nd Layer: 15 Nurons
model.add(Dense(15,activation='relu'))

#OUTPUT LAYER: 1 NURON
model.add(Dense(1,activation='sigmoid'))

#COMPILE Model
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
#EARLY STOPPING CALLBACKS
from tensorflow.keras.callbacks import EarlyStopping

help(EarlyStopping)
#So we have to choose a Metric to MONITOR, in our case it 
# will be VALIDATION LOSS. So we will track VAL LOSS. You can
# also specify things like MINIMUM CHANGE REQUIRED. Also 
# theres PATIENCE which is the number of EPOCHS with NO 
# IMPROVEMNET after which TRAINING will be STOPED. So it 
# deosnt' just do a HARD STOP the minute the VAL LOSS goes 
# up a little, so maybe we wanna wait a couple of EPOCHS. 
```

```python
#EARLY STOPPING
#Monitor:ValLoss
#Remeber LOSS=0=PERFECT FIT
#Mode: MIN. Recall the MODE is what you are trying to do. 
# So are you trying to minimize the thing you are monitoring 
# or are you trying to maximize the thing you are monitoring
# ect? So if your metric was accuracy, you would want to 
# maximize it. If you metric is LOSS, then you would want to
# MINIMIZE it
 #There's also an AUTO STRING for MODE and direction is 
#     automatically inferred from name of the moinitored 
#     quantity. This usually almost always works, but 
#     recommended to put something other than AUTO for it  
#PATRIENCE=25, so that means we will wait 25 EPOCHS even 
 #after we have detected a STOPPING pt bc of the noise that
    #could occur
early_stop = EarlyStopping(monitor='val_loss',mode='min',
                          verbose=1,patience=25)
```

```python
#FIT MODEL again
#add CALLBACK pararm as a LIST
model.fit(x=X_train,y=y_train,epochs=600,
          validation_data=(X_test,y_test),
          callbacks=[early_stop])
#This will attempt to run on 600 EPOCHS UNLESS EARLY STOP
 #is triggered
#We see EARLY STOPP was triggerd and it STOPED TRAINING 
 #at 91
```

```python
#So whats nice about this is we can choose a LARGE EPOCH
#and then indicate you want it to stop early with CALLBACK
```

```python
#Check MODEL LOSS
model_loss = pd.DataFrame(model.history.history)
model_loss
```

```python
#PLOT MODEL LOSS
model_loss.plot()
#This is exatly the kind of plot we want to see where LOSS
# and VAL LOSS are both begining to go down and right as they
# begin to spread out if probly a good indication of where we
# should stop training. Notice the VAL LOSS starts to FLATTEN
# out which is OKAY. What we want to prevent is VAL LOSS 
# beginning to INCREASE and thats what happened so EARLY 
# STOP was triggered
```

```python
#3rd thing we can do to PREVENT OVERFITTING:
 #ADD DROPOUT LAYERS
#DROPOUT LAYERS will TURN OFF a PERCENTAGE of NURONS 
 #randomly. 
from tensorflow.keras.layers import Dropout
```

```python
#REDO MODEL
#AFTER each of the 2 LAYERS, we will add in DROPOUT CALL
#RATE is the probability that you are going to RANDOMLY turn 
# off the actual NURONS. So its the FRACTION of which you are
# randomly turning off. A 0 means you are taking 0 percent of
# the NURONS and randomly turning them off during training. 
# If you put a 1 it would mean 100% of the NURONS will be 
# turned off randomly for each batch of training. Its common
# to use something between 0.2 and 0.5. and that means HALF 
# the NURONS during EACH BATCH(essentially each batch is 1 
# ENTIRE EPOCH of training in our case since we haven't
# specified a batch size) in this layer
# of 30 are going to be turned off so their weights and biases
# wont be getting updated

model = Sequential()

#1st Layer: 30 Nurons. So half will be turned off RANDOMLY
model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))
#2nd Layer: 15 Nurons. So half will be turned off RANDOMLY
model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))

#OUTPUT LAYER: 1 NURON
model.add(Dense(1,activation='sigmoid'))

#COMPILE Model
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
#USE SAME EARLY STOPPING WE DEFINED ABOVE
```

```python
#RE-FIT MODEL
model.fit(x=X_train,y=y_train,epochs=600,
          validation_data=(X_test,y_test),
          callbacks=[early_stop])
#Stopped at 143 EPOCHS
#we notice it ran LONGER and that is GOOD bc, that means it
#is still learning on those further EPOCHS, and thats due to
#the added DROPOUT LAYERS
```

```python
#Analyze LOSSES
model_loss = pd.DataFrame(model.history.history)
model_loss
```

```python
#Plot MODEL LOSS
model_loss.plot()
#This is EVEN BETTER this is FANTASTIC. TRAIN LOSS and 
# VAL LOSS are both quickly going down and they are flattening
# out at the same rate. This is eactly want we want to see and
# much improved. 
```

```python
#So adding in EARLY STOPPING and DROPPOUT LAYERS has 
#significantlly increased our performance
```

```python
#Lets do a FULL EVAULATION on our CLASSES
```

```python
#PREDICTIONS 
 #method is a little differnet since we are doing a 
     #CLASSIFICATION task
predictions = model.predict_classes(X_test)
predictions
#we get CLASSES it PREDICTS for the X TEST data set
#Recall even if you had a new tumor with all these features
# it still just predicts classes. A new datapoint is 
# essentially the exat same thing as a TEST SET 
```

```python
#CLASSIFICATION RPT/CONFUSION MATRIX
from sklearn.metrics import classification_report,confusion_matrix

#Compare Y_TEST vs PREDICTIONS 
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
#so we have VERY good performance, our network only 
#misclassified 1 point
#97% ACC
#ACCURACY is a good METRIX bc we saw we had a relatively
#balances classes. they weren't perfect but weren't 
#extremely unbalanced either
```
