```python
# OBJECT ORIENTED METHOD - more control over plotting
   # main idea in using this is to create figure objects and
       #call methods off of this
```

```python
#Create figure object, then add axes, then plot
fig = plt.figure()
# this is basically an imaginary blank canvas and we can add
   # a set of axis to this canvas
# pass in a list to creat a set of axes 
    # 4 arguments for axes: 1-Left,2-Bottom,3-Width,4-Height
        # These range from 0-1 (so it's a percent of canvas we 
           # want to take)
axes = fig.add_axes([0.1,0.1,0.8,0.8])
# Next we can plot on thte set of axes: familiar to above
axes.plot(x,y)
```

```python
# Add X labels,Ylabels, and Titles
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
# OBJECT ORIENTED METHOD 2
  # this method for subplots that you want arranged by rows 
    #columns. Allows you to tuple unpack (fig,axes) = subplots
    # call. And based off those axes we get an array of axes
    # and we can plot or set titles by INDEXING or by 
    # ITERATING through the axes OBJECT
```

```python
# OBJECT ORIENTED METHOD 2

# Create Subplots using Object Oriented Method
# instead of calling figure, since we are creating subplots
    # call subplots
# SUBPLOTS DIFFERS BC YOU CAN SET NUMBER OF ROWS AND COLUMNS
# fig,axis = .... DOING TUPLE UNPACKING
fig,axes = plt.subplots()

# use axes object to plot on it
axes.plot(x,y)
```

```python
# SUBPLOTS DIFFERS BC YOU CAN SET NUMBER OF ROWS AND COLUMNS
# ALL SUBPLOTS ARE DOING IS CALLING THE FIG.ADD_AXES 
   #AUTOMATICALLY FOR YOU
fig,axes = plt.subplots(nrows=1,ncols=2) 
# SO WE HAVE 1 ROW WITH 2 COLUMNS
```

```python
axes
# Axes is an array LIST of MatPlotLib axes - meaning we can 
   #ITERATE THROUGH IT
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
```

```python
#Fix the Layout - TIGHT_LAYOUT
fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
# ITERATE THROUGH THE AXES
fig,axes = plt.subplots(nrows=1,ncols=2) 

for current_ax in axes:
    current_ax.plot(x,y)
# THIS JUST SHOW THAT AXES OBJECT IS JUST A LIST OF MATPLOTLIB
   #AXES WHICH ARE THE ONES WE MANUALLY CREATED WHEN WE SAID 
   #FIG.ADD_AXES
# ONLY DIFFERENCE BETWEEN PLT.SUBPLOTS AND JUST DOING 
   #PLT.FIGURE IS THAT PLT.SUBPLOT IS JUST DOING ADD_AXES 
    # STUFF FOR YOU AUTOMATICALLY BASED OFF HOW MANY ROWS AND
       #COLUMNS YOU WANT
```

```python
# Since we can ITERATE thru axes object we can also INDEX it
fig,axes = plt.subplots(nrows=1,ncols=2) 

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
# LEGENDS 
  # shows which plot is what 
  # can use label text to clarify what plot is what plot
  # 2 STEP PROCESS 
       #1- for each PLOT CALL - add in argument call Label
       #2- ax.legend()
```

```python
fig = plt.figure()
                 
ax = fig.add_axes([0,0,1,1])   

ax.plot(x,y)
ax.plot(y,x)
```

```python
# Add a LEGEND
fig = plt.figure()
                 
ax = fig.add_axes([0,0,1,1])   

ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend()
```

```python
# Can specify Legend location using LOC
  # for LOC: specify: location code and location string
   # recommend location code = 0 for best location from MPL
# Location codes are 0-10 (see documentation)
fig = plt.figure()
                 
ax = fig.add_axes([0,0,1,1])   

ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend(loc=0)
```

```python
# If LOC doesn't work we can specify a 2-tuple giving x,y of 
  #the lower left corner of the legend in axes coordinated
# LOC using TUPLE arrangement 
fig = plt.figure()
                 
ax = fig.add_axes([0,0,1,1])   

ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend(loc=(0.1,0.1))
```

```python
#Settting Colors, Linewidths, Linetypes, and other Customizing 
   #Options of how plot looks 
```

```python
# Setting Colors
# Color argument can take in common color strings like blue or
   #RGBD hexcodes
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00')
```

```python
# Linewidth and Linestyle
# 1 is default linewidth  (lw shorthand)
# Another useful tip for linewidth and visibily is the alpha
   # argument 
# Alpha argument controls how transparent the line is
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=3)
```

```python
# Alpha 
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=3,alpha=0.5)
```

```python
# Linestyle  -     :,--,-.,.,steps       (ls shorthand)
  # defines style of line 
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=3,linestyle='--')
```

```python
# Lets say we want to add MARKERS where X points occur on plot
# Many avaiable :   o,+,*,1(just a num code)
# Markersize - used to make marker bigger
# Many options to specify what marker looks like:
   #markerface color, markeredge width, markeredge color
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=1,ls='-',marker='o',markersize=20)
```

```python
# Edit what marker looks like
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=1,ls='-',marker='o',markersize=20,
       markerfacecolor='yellow',markeredgewidth=3,
       markeredgecolor='green')
```

```python
# Control over axis appearance and plot range
# Set Y-limit and X-limit
# Only want to show plot between 0 and 1 on X-axis
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
